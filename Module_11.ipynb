```python

import seaborn as sns
from dash import Dash,html, dcc, input, output, callback, dash, table
df = sns.load_dataset("tips")

app = Dash()

app.layout = html.Div(children =[
    html.H1("choose your option"),
    dcc.slider(df['tip'].max(), value = df['tip'].mean(),id = "my slider", tooltip = true),
    dash_table.DataTable(id="dt")
])

@callback(output("dt","data"),Input("my_slider", "value"))
def dropdown_selection(user_input):
    temp_df = df[df["tip"]<=user_input]
    return temp_df.to_dict('records')
app.runserver()
```
